```python
import pandas as pd

df = pd.read_csv("../data/cleaned/cleaned_data.csv")
```

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)
```

```python
total_orders = df['Order ID'].nunique()
print("Total Orders:", total_orders)
```

```python
aov = total_sales / total_orders
print("Average Order Value:", aov)
```

**Insight:** The business generated $2261536.78 in total revenue across 4922 orders, with an average order value of $459.47.

```python
yearly_sales = df.groupby('Year')['Sales'].sum()
yearly_sales.plot()
```

```python
monthly_sales = df.groupby('Month Name')['Sales'].sum()
monthly_sales
```

**Insight:** “Sales show a peak during Q4, indicating strong seasonal demand.”

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales
```

**Insight:** “Technology is the highest revenue-generating category.”

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
subcat_sales.head(10)
```

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)
top_customers.head(10)
```
